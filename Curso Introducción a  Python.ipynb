**PREGUNTA: ¿Qué hay de diferencia entre una VISTA y una COPIA?**

Está muy bien definido en la documentación del curso, pero en definitiva, cuando se trabaja sobre una **VISTA** es importante recordar que se ***modifica los datos de esa variable original***, mientras que una **COPIA**, como su nombre indica, los cambios se realizarían sobre esa segunda variable creada a partir de una primera, y de esta forma, ***no se modifica el valor que tiene la variable original***.

Solución ***missing values***.

```python
print('---- Datos faltantes ----')
print(dataFrame.isna().sum())
print('-------------------------\n')

print('\n-------------- Nuevo DataFrame sin datos faltantes --------------\n')
dFrame = dataFrame.dropna()
print(dFrame)
```

Cálculo de la **media** y **desviación típica** de las variables numéricas.

```python
"""
El cálculo de la media y la desviación típica para las variables CustomerID y CustomerCityID no tiene sentido, ya que,
su valor es un dato identificativo único, el primero para cada cliente y el segundo para cada ciudad.

Sin embargo, ambos cálculos para las variables CustomerAge, CreditScore y SpendingScore, sí que puede ser un dato relevante.
"""
dF_paraCalculos = dFrame.copy()

try:
  dF_paraCalculos.drop(['CustomerID', 'CustomerCityID'], axis='columns', inplace=True)
except KeyError as ex:
  print('Columna ya eliminada del DataFrame.\n')

print('--- Media de cada variable numérica ---')
print(dF_paraCalculos.mean())
print('---------------------------------------\n')

print('\n-------- Media por CustomerAge --------')
print(dF_paraCalculos.groupby('CustomerAge').mean())
print('-----------------------------------------\n')
```

```python
print('--- Desviación típica de cada variable numérica ---')
print(dF_paraCalculos.std())
print('---------------------------------------------------\n')

print('\n--- Desviación típica por CustomerAge ---')
print(dF_paraCalculos.groupby('CustomerAge').std())
print('-------------------------------------------\n')
```

Crea **tres funciones** por separado para sumar, restar y multiplicar.

```python
# 1. Crea tres funciones por separado para sumar, restar y multiplicar.

def sumar(a,b):
  r = a + b
  return r

def restar(a,b):
  r = a - b
  return r

def multiplicar(a,b):
  r = a * b
  return r


# 2. Pregúntale al usuario qué operación le gustaría realizar, dándole un número a cada una de ellas (p.ej: 1 = Restar)

op = int(input("Seleccione el número de la operación que quiere realizar: \n"
            + "1. Sumar - 2. Restar - 3. Multiplicar: "))


# 3. Crea un BUCLE IF para ejecutar la operación matemática, donde primero le pida al usuario los dos números a sumar/restar/multiplicar.

if op == 1:
  a = int(input("Introduzca el primer número para la suma: "))
  b = int(input("Introduzca el segundo número para la suma: "))
  print("Resultado: " + str(sumar(a,b)))
elif op == 2:
  a = int(input("Introduzca el primer número para la resta: "))
  b = int(input("Introduzca el segundo número para la resta: "))
  print("Resultado: " + str(restar(a,b)))
elif op == 3:
  a = int(input("Introduzca el primer número para la multiplicación: "))
  b = int(input("Introduzca el segundo número para la multiplicación: "))
  print("Resultado: " + str(multiplicar(a,b)))
else:
  print("Introduzca un número válido")
```
