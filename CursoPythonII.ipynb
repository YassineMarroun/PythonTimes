**Web scraping**

Obtenemos los datos una URL y procemos los datos gracias a Beautiful Soup, que se trata de una librería de Python.


```
# Se importan los módulos
import requests
from bs4 import BeautifulSoup

# Dirección de donde obtenemos la información
url = "https://www.atptour.com/es/rankings/singles"

# Se realiza el GET request de la dirección url
response = requests.get(url)
# Se analiza el texto html obtenido
html = BeautifulSoup(response.text, 'html.parser')

# Listas donde guardaremos los nombres y las edades de los jugadores
nombres = []
edades = []


# Se recorre buscando todas las etiquetas td con la clase "player-cell border-left-dash-1 border-right-dash-1"
for item in html.find_all("td", class_="player-cell border-left-dash-1 border-right-dash-1"):
    n = ""

# Por cada etiqueta encontrada, buscamos dentro una etiqueta span con la clase "player-cell-wrapper" para el nombre del jugador
    for item2 in item.find_all("span", class_ = "player-cell-wrapper"):
      if item2.text != "": n = item2.text.replace("\r\n", "").strip()
      # Guardamos todos los nombres de jugadores en una lista
      nombres.append(n)


# Buscamos dentro de toda etiqueta td que tenga la clase "age-cell border-left-dash-1 border-right-4" para obtener la edad
e = ""

for item3 in html.find_all("td", class_ = "age-cell border-left-dash-1 border-right-4"):
  if item3.text != "": e = item3.text.replace("\r\n", "").strip()
  # Guardamos todas las edades de los jugadores en una lista
  edades.append(e)

# Pasamos las dos listas a un diccionario donde el nombre será la clave y la edad el valor
jugadores = dict(zip(nombres, edades))

print(jugadores)


# A partir de aquí, podríamos jugar todo lo que quisiesemos con el diccionario de jugadores que hemos creado.
# Por ejemplo, vamos a recorrer el diccionario y mostrar diferentes listados en función de la condición que le digamos.


# 1. Mostrar el top 10 mundial del ranking ATP.
print('\n Top 10 mundial de la ATP')
print('---------------------------')
numero = 1
for nombre in jugadores.keys():
  if numero < 11:
    print(str(numero) + ': ' + nombre)
  numero += 1


# 2. Mostrar todos los jugadores menores de 21 años junto con su posición en el ranking mundial
print('\n Jugadores menores de 21 años en el top 100 mundial de la ATP')
print('---------------------------------------------------------------')
numero = 1
for nombre in jugadores.keys():
  edad = int(jugadores.get(nombre))
  if edad <= 21:
    print(str(numero) + ': ' + nombre + ' --> Edad: ' + str(edad))
  numero += 1
```
